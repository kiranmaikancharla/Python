```python
# Read a number as input and print the output as Factors of the given number

# Input : 6
# Output : 1 2 3 6

# Input : 12
# Output : 1 2 3 4 6 12

def factors(n):
    i = 1
    for i in range(1,n+1):
        if n % i == 0:
            print(i,end=' ')
            
print(factors(6))
print(factors(12))
```

```python
# Read the input as Number and print the output as True or False. 
# If the input number is Prime True otherwise print False

# Input : 11
# Output : True

# Input : 10
# Output : False

def prime(n):
    if n > 1:
        for i in range(2,n):
            if n % i == 0 :
                print("False")
                break
        else :
            print("True")

prime(11)
prime(10)
```

```python
# Read the input as num N and Print the output as Prime number count btw 1 to N

# Input : 10
# Output : 4(2,3,5,7)

def countPrime(l,u):
    count = 0
    for n in range(l,u+1):
        if n>1:
            for i in range(2,n):
                if n % i == 0:
                    break
            else :
                print(n,end=' ')
                count = count + 1
    print("count is :",count)
countPrime(1,10)
```

```python
# Read the i/p as two limits,Print the o/p as Palindrome count bet the 2 limits

# Input : 1 10 
# Output : 9

# Input : 1 100
# Output : 18

def palindrome(s,n):
    count = 0
    for i in range(s,n+1):
        temp = i
        rev = 0
        while temp>0:
            r = temp % 10
            rev = rev * 10 + r
            temp = temp // 10
        if i == rev:
            count = count+1
            print(i)
    print("count is: ",count)   
palindrome(1,10)        
```

```python
# Read the  i/p as number, 
# Print the o/p as Count of triangles and Circles in the given num

# Input : 12854
# Output : 3
# Input : 888
# Output : 6

# Reason : 0,6,9 wil have one circle so count wil be 1
# 4 wil have one triangle so count wil be 1
# 8 wil have 2 circles so count wil be 2

def count(n):
    s = 0
    r = 0
    while n != 0 :
        r = n % 10
        if r == 0 or r == 6 or r == 9:
            s = s + 1
        elif r == 4 :
            s = s + 1
        elif r == 8 :
            s = s + 2
        n = n // 10
    return(s)
print(count(12854))
print(count(888))
        
```
