#### Visualizing Data
- An intresting way to display Dataframe
- With matplotlib

```python
import matplotlib.pyplot as plt
a1 = [1,2,3,4]
a2 = [5,6,7,8]
plt.figure(figsize=(4,4))
plt.plot(a1,a2,color='green')
plt.xlabel('x-Axis')
plt.ylabel('y-Axis')
plt.show()
```

```python
divisions = ['A','B','C','D']
marks = [77,45,67,87]
avg_marks = [49,78,98,87]
width = 0.70
plt.bar(divisions,marks,color='blue',width = 0.3)
plt.show()
```

```python
plt.barh(divisions,marks,color='blue')
plt.show()
```

```python
lang = ["python","java","c","c++"]
rating = [75,70,80,90]
plt.pie(rating,labels = lang,shadow=True,startangle=90,autopct="%1.1f%%")
plt.legend(loc="best")
plt.show()
```
